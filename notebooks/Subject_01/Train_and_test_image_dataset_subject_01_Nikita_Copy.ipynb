## Save test and train images in folders

```python
test_data_indices
```

```python


# Folder path where you want to save the images
folder_path = "../../data/test_images_indexed"

# Assuming you have a list of NumPy arrays named "image_arrays"
image_arrays = X_test_image_data
image_indices = test_data_indices
# Create the folder if it doesn't exist
if not os.path.exists(folder_path):
    os.makedirs(folder_path)

# Save each array as an image
for index, image_array in zip(image_indices, image_arrays):
    image_path = os.path.join(folder_path, f"{index}.png")
    image = Image.fromarray(image_array)
    image.save(image_path)
```
