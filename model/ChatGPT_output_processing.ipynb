```python
text_list = []
unique_id_list = []

with open(filename,'r') as train_output:
    
    
    for line in train_output:
            
        if line.split()[0] == '"text":':
            
            text_list.append(line.split()[1:])
```

```python
X_train_dict = {}

for text, unique_id in zip(text_list,unique_id):
    
    X_train_dict[unique_id] = text
```

```python
for key in X_train_dict.keys():
    
    text_to_clean = X_train_dict[key]
        
    new_text = [text.replace("\\n","").replace('"','').replace(',',"").replace('\\','').replace('`','').lower() for text in text_to_clean]
    
    
    X_train_dict[key] = new_text
```

```python
lem = WordNetLemmatizer()
```

```python
four_value_X_train = {}

for key, value in X_train_dict.items():
    value_list = []
    val_list_2 = []
    
    vlemmatized = [lem.lemmatize(word, 'v')
                       for word in value]
    for word in vlemmatized:
        value_list.append(word)
    nlemmatized = [lem.lemmatize(word, 'n')
                       for word in value_list]
    for word in nlemmatized:
        val_list_2.append(word)
#     nlemmatized
    if len(val_list_2) == 2:
        val_list_2.append(" ")
    
    if len(val_list_2) == 3:
        val_list_2.append(" ")
    
    for idx, word in enumerate(val_list_2):
        if len(word) <= 2 and word not in two_character_words:
            val_list_2[idx] = " "
    
    four_value_X_train[key] = val_list_2
```
