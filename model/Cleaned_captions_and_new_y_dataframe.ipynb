```python
def remove_punc(text):
    for punctuation in string.punctuation:
        text = text.replace(punctuation, '')
    return text
```

```python
from nltk.corpus import stopwords 
stop_words = set(stopwords.words('english'))
stop_words
```

```python
stopwords_removed_train = []

for caption in train_no_punc:
    caption_without_stopwords = [w for w in caption.lower().split() if not w in stop_words]
    stopwords_removed_train.append(caption_without_stopwords)
    
    
    
```

```python
stopwords_removed_train
```

```python
stopwords_removed_train[1]
```

```python
lem = WordNetLemmatizer()
```

```python
numbers_to_remove = ['104', '9', '5', '12', '120', '122', '150', '18th', '252', '3', '350', '360'
                    '3D', '2', '4', '50', '52', '535', '550', '60th', '65', '66520', '700', '95', '99', '3d', '360']
```

```python
words_to_remove = ['ha', 'go', 'dc', 'bu', 'sa', 'le', 'fa', 'sk', 'k', 'nd', 'w', 'u', 'aq', 'st', 'vw', 'us',
                   'be', 'ex', 'ok', 'b5', 'ft', 'el', 'p', 'x', 're', 'os', 'an', 'th', 'do', 'l']
```

```python
lem_train = []

for caption in stopwords_removed_train:
    value_list = []
    val_list_2 = []
    
    vlemmatized = [lem.lemmatize(word, 'v')
                       for word in caption]
    for word in vlemmatized:
        value_list.append(word)
    nlemmatized = [lem.lemmatize(word, 'n')
                       for word in value_list]
    for word in nlemmatized:
        val_list_2.append(word)
        
    for idx, word in enumerate(val_list_2):
        if word in numbers_to_remove: 
            val_list_2[idx] = ''
            
    for idx, word in enumerate(val_list_2):
        if word in words_to_remove:
            val_list_2[idx] = ''
        
    lem_train.append(val_list_2)
    
    
```
